```python
import requests as re 
import os
import pandas as pd
```

```python

arquivo = pd.read_csv(os.getcwd()+os.sep+'acoes-listadas.csv',sep= ',',index_col=False)
```

```python
empresas = arquivo['Nome']
```

```python
if str(re.get('https://logo.clearbit.com/Lojas Renner.com.br').content) == "b'\\n'" or 'not a valid' in str(re.get('https://logo.clearbit.com/Lojas Renner.com.br').content):
    print('ha erro')
```

#### API Content Request -> Logo Crearbit 

```python
for empresa in empresas:
    arquivo = re.get('https://logo.clearbit.com/{}.com'.format(empresa)).content
    if str(arquivo) == "b'\\n'" or 'not a valid' in str(arquivo):
        arquivo = re.get('https://logo.clearbit.com/{}.com.br'.format(empresa)).content
        if str(arquivo) == "b'\\n'" or 'not a valid' in str(arquivo):
            empresa =''
            arquivo = ''
        print(empresa)
    if arquivo != '':
        with open('imagens\{}.png'.format(empresa), 'wb') as f:
            try:
                f.write(arquivo)
            except:
                pass
```
